## Theory
### Why normalize the database?

I would separate the data into Orders, Customers, Products, and Date tables to reduce redundancy, improve consistency, and simplify maintenance. Relationships between these tables would be created using primary and foreign keys.

### Why change indexing for 500 million rows?

I would create indexes on frequently filtered or joined columns, such as OrderID, CustomerID, ProductID, and OrderDate. While indexes improve SELECT performance, they can slow down INSERT and UPDATE operations because the indexes must also be updated.

## Theory
### Why Star Schema?

A star schema separates the fact table from dimension tables, making queries faster and the data model easier to understand.

### Why a Date Table?

A dedicated Date table enables Power BI time intelligence functions such as YTD, MTD, QTD, and MoM calculations.

### Measures vs Calculated Columns

Measures are calculated dynamically based on the current filter context, making them ideal for KPIs and reports. Calculated columns are computed during data refresh, stored in the model, and are best for creating row-level attributes.

### Automatic Refresh

I would publish the report to Power BI Service, configure Scheduled Refresh, and, if using an on-premises SQL Server, install an On-premises Data Gateway.
